## Scope
The New York's Marathon is considered one of the most challenging races around the world. As it is one the 6 major marathons in the world, it is one the main events in the city every year.

The tablue file shared on link below includes different visualizations to analise the behaviour of the New York Citi Bike usage on  the day of the New York Marathon for 2022. The dashboards and Story pretend to compare what is the difference between a normal Sunday compare to thr Sunday of the NYC Marathon

The was developed on Tableau and is public on the following link:
https://public.tableau.com/views/Challenge_Tableau_17009086779720/NYCMarathon?:language=en-GB&publish=yes&:display_count=n&:origin=viz_share_link

## Results
- One of the main conclusions of the analisis is that the New York Citi Bike shows an increment of 47% the day of the NYC Marathon. Therefore, it is needed to increment the availability of bikes and options the day of the race to supply that demand.

- As it was expected, the nunmber of casual users move from 20% on a normal day, to 30% on the day of the race, showing the nunmber of turists or foreigners that use the system the day of the race. It would be relevant to include some special deals or sales for causal members the day of the Marathon.

- The line chart shows a similar behaviour for the race day and the next Sunday. That shows that the city keeps using the system in the same way independntly of the race. Nevertheless, the frequency is double for the hours of the race (betwwen 9am and 3pm)

```python
data_202211['started_at'] = pd.to_datetime(data_202211['started_at'])
data_202211['ended_at'] = pd.to_datetime(data_202211['ended_at'])
data_202211['day'] = data_202211["started_at"].dt.day_name()
```

```python
data_202211['duration'] = data_202211['ended_at']-data_202211['started_at']
data_202211['minutes'] = data_202211['duration'] / pd.Timedelta(minutes=1)
                             
data_202211.tail(10)
```

```python
data_202211 = data_202211.loc[data_202211["minutes"]>=1].sort_values(by=['started_at'])
data_202211 = data_202211.loc[((data_202211["start_lat"]) != (data_202211["end_lat"])) & 
                              ((data_202211["start_lng"]) != (data_202211["end_lng"]))]
data_202211 = data_202211.reset_index(drop=True)
print(len(data_202211.index))
data_202211.tail()
```

```python
data_202211_1 = data_202211.loc[data_202211["started_at"] <= '2022-11-10 00:00:00']
data_202211_2 = data_202211.loc[(data_202211["started_at"] > '2022-11-10 00:00:00') & 
                                (data_202211["started_at"] <= '2022-11-20 00:00:00')]
data_202211_3 = data_202211.loc[data_202211["started_at"] > '2022-11-20 00:00:00']
data_202211_3.tail()
```

```python
# Exporting to csv files
data_202211_1.to_csv("Resources/raw_data_1.csv")
data_202211_2.to_csv("Resources/raw_data_2.csv")
data_202211_3.to_csv("Resources/raw_data_3.csv")
```
